### **Task 1**: 
Plot the distribution of combined fuel mileage (column 'comb', in miles per gallon) by manufacturer (column 'make'), for all manufacturers with at least eighty cars in the dataset. Consider which manufacturer order will convey the most information when constructing your final plot. 

**Hint**: Completing this exercise will take multiple steps! Add additional code cells as needed in order to achieve the goal.

```python
manuf = fuel_econ.make.value_counts().loc[lambda x: x >= 80].index.tolist()
type(manuf)
```

```python
len(manuf)
```

```python
df = fuel_econ.loc[fuel_econ.make.isin(manuf)]
```

```python
# manuf_class = pd.api.types.CategoricalDtype(ordered=True, categories=manuf)
# df['make'] = df['make'].astype(manuf_class)

make_means = df.groupby('make').mean()
comb_order = make_means.sort_values(by='comb', ascending=False).index
```

```python
df.comb.describe()
```

```python
bins = np.arange(13, 58+2, 2)
g = sb.FacetGrid(data=df, col='make', col_wrap=6, height=3, col_order=comb_order) # 'height': the title size of subplots
g.map(plt.hist, 'comb', bins=bins)
g.set_xlabels('comb') # Label the x axis on the bottom row of the grid.
g.set_titles(col_template='{col_name}'); # Draw titles either above each facet or on the grid margins.
```

```python
bins = np.arange(13, 58+2, 2)
g = sb.FacetGrid(data=df, col='make', col_wrap=6, height=3, col_order=comb_order)
g.map(sb.distplot, 'comb', bins=bins) # use 'sb.distplot' to see the plot in terms of density
g.set_xlabels('comb')
g.set_titles(col_template='{col_name}');
```

Continuing on from the previous task, plot the mean fuel efficiency for each manufacturer with at least 80 cars in the dataset.

```python
base_color = sb.color_palette()[0]
sb.barplot(data=df, x='comb', y='make', color=base_color, ci='sd', order=comb_order)
plt.xlabel('Avg. Combined Fuel Efficiency (mpg)');
```
